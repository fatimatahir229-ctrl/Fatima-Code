# **APPLE MOBILE ANALYSIS**

```python
import pandas as pd
```

```python
df = pd.read_csv("apple_products.csv")
```

```python
df["Product Name"]
```

```python
result = {}

for item in df["Product Name"]:
    name = item.split("(")[0].strip()
    model = item.split("(")[1].replace(")"," ")

    phone = model.split(",")
    color = phone[0].strip()

    if len(phone) > 1:
        storage = phone[1].strip()
    else:
        storage = None

    if name not in result:
        result[name] = {"Colors": [], "Storages": []}

    if color and color not in result[name]["Colors"]:
        result[name]["Colors"].append(color)
    if storage and storage not in result[name]["Storages"]:
        result[name]["Storages"].append(storage)


for name, details in result.items():
    print(f"{name} Colors: {(details['Colors'])} Storages:{(details['Storages'])}")

```

```python
new_rows = []
for name, details in result.items():
    new_rows.append({
        "Product Name": name,
        "Available Colors": ", ".join(details["Colors"]),
        "Available Storages": ", ".join(details["Storages"])
    })

result_df = pd.DataFrame(new_rows)

result_df.to_csv("Apple.csv")

```
