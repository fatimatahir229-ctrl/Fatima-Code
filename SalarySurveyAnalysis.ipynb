# **Salary Survey Analysis**

```python
import pandas as pd
dfSal = pd.read_excel("ManagerSurvey.xlsx")
```

```python
dfSal
```

```python

dfSal=dfSal.set_axis(['TimeStamp','Age Range','Industry','Job Title',
                'Title Context','Net Salary','Salary Bonus','Currency',
                'Other Currency','Additional Income Context','Country of Work','US State',
                'Work City','Work Experience Overall','Work Experience in your field',
                'Highest Education','Gender','Race'],axis=1)
```

```python
dfSal['Age Range'].unique()
```

```python
#idmax() is used to tell name of variable having highest count
#max() tells the count of that variable
count_maxage=dfSal['Age Range'].value_counts().idxmax()
count_1=dfSal['Age Range'].value_counts().max()
count_minage=dfSal['Age Range'].value_counts().idxmin()
count_2=dfSal['Age Range'].value_counts().min()
print(f'Age range having maximum responses: {count_maxage} ({count_1})')
print(f'Age range having minimum responses: {count_minage} ({count_2})')
```

```python
AllAges=[18,24,25,34,35,44,45,54,55,64,65]
Meanage=sum(AllAges)/len(AllAges)
print('Mean Age: ',Meanage)
```

```python
dfSal['Currency'].unique()
```

```python
CurrencyRates={'USD':1, 'GBP':1.33, 'CAD':0.73, 'EUR':1.10, 'AUD/NZD':0.64, 'Other':1, 'CHF':1.10, 'ZAR':0.052,
       'SEK':0.092, 'HKD':0.128, 'JPY':0.0068}
```

```python
dfSal['Net Salary'].info()
```

```python
dfSal['Net Salary'] = dfSal['Net Salary'].astype(float)

for i in range(len(dfSal)):
    curr = dfSal.loc[i, 'Currency']

    if curr == 'GBP':
        dfSal.loc[i, 'Net Salary'] *= CurrencyRates['GBP']
    elif curr == 'CAD':
        dfSal.loc[i, 'Net Salary'] *= CurrencyRates['CAD']
    elif curr == 'AUD/NZD':
        dfSal.loc[i, 'Net Salary'] *= CurrencyRates['AUD/NZD']
    elif curr == 'CHF':
        dfSal.loc[i, 'Net Salary'] *= CurrencyRates['CHF']
    elif curr == 'ZAR':
        dfSal.loc[i, 'Net Salary'] *= CurrencyRates['ZAR']
    elif curr == 'SEK':
        dfSal.loc[i, 'Net Salary'] *= CurrencyRates['SEK']
    elif curr == 'HKD':
        dfSal.loc[i, 'Net Salary'] *= CurrencyRates['HKD']
    elif curr == 'JPY':
        dfSal.loc[i, 'Net Salary'] *= CurrencyRates['JPY']
    else:
        dfSal.loc[i, 'Net Salary'] *= 1
```

```python
datafram=dfSal[dfSal['Country of Work'].isin(['US', 'USA', 'U.S.', 'United States'])]
```

```python
datafram.loc[:, 'Country of Work'] = datafram['Country of Work'].replace(
    ['US', 'USA', 'U.S.', 'United States'], 'US'
)
```

```python
datafram
```

```python
max_country=datafram['Country of Work'].value_counts().idxmax()
max_country_count=datafram['Country of Work'].value_counts().max()
if 'US State' in datafram.columns:
    max_state=datafram['US State'].value_counts().idxmax()
    max_state_count=datafram['US State'].value_counts().max()
print(f'country having maximum responses are: {max_country}  ({max_country_count})\nstate having maximum responses are: {max_state}  ({max_state_count})')

top_state=datafram[datafram['US State'] == max_state]
if 'Work City' in top_state.columns:
    max_city=top_state['Work City'].value_counts().idxmax()
    max_city_count=top_state['Work City'].value_counts().max()
print(f'city having maximum responses are: {max_city}  ({max_city_count})')
```

```python
top_city=top_state[top_state['Work City']==max_city]
if 'Gender' in top_city.columns:
    gender1=top_city['Gender'].value_counts()

print(gender1)
    
```
