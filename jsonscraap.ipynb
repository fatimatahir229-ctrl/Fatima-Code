# **JSON SCRAPING**

```python
import requests as r
import pandas as pd
readingsite = r.get("https://dummy-json.mock.beeceptor.com/continents")
with open("Continents.csv", "w", encoding="utf-8") as f:
    f.write(readingsite.text)

with open("Continents.csv", "r", encoding="utf-8") as f:
    content = f.read()

companiesUrlResponses = r.get("https://dummy-json.mock.beeceptor.com/continents")
companiesResponse = companiesUrlResponses.json()
company = {
    "code": [],
    "name": [],
    "areaSqKm": [],
    "population": [],
    "countries": [],
    "developedCountries": []
}

for item in companiesResponse:
    company["code"].append(item.get("code"))
    company["name"].append(item.get("name"))
    company["areaSqKm"].append(item.get("areaSqKm"))
    company["population"].append(item.get("population"))
    company["countries"].append(item.get("countries"))
    company["developedCountries"].append(item.get("developedCountries"))

df = pd.DataFrame(company)
print(df)
df.to_csv("Continents.csv")
```

# **MORE EFFICIENT WAY**

```python
import requests as r
from pandas import DataFrame as df
companiesUrlResponse = r.get("https://dummy-json.mock.beeceptor.com/continents")
companiesResponse = companiesUrlResponse.json()
continentsDictionary = {}
for di in companiesResponse:
    for element in di.keys():
        continentsDictionary[element] = []
for di in companiesResponse:
    for element in di.keys():
        continentsDictionary[element].append(di[element])

dfcontinents = df(continentsDictionary)
print(dfcontinents)
dfcontinents.to_csv("continents.csv")
```
